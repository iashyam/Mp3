````python
v0 = 10 # intial potential guess
N_max = 100 #maximum number of points in the aqure gird
U = np.zeros([N_max, N_max]) #let the initial guess be 0
for i in range(0,N_max-1): #putting the intial condions that first x=0 par U 100V hai
    U[i][0] = 100
````

````python
for _ in range(400): #maximum iteration is 100
    for i in range(N_max-1):
        for j in range(N_max-1):
            U[i][j] = (1/4)*(U[i-1][j]+ U[i][j-1]+U[i+1][j]+U[i][j+1])
````

````python
plt.plot(U[2:-2]);
````

````python
x, y = np.meshgrid(np.linspace(0,1,N_max),np.linspace(0,1,N_max))
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,U)
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Potential')
````
